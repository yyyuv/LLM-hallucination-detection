```python
token_to_word_index_mappings = []

for i, description in enumerate(raw_descriptions):
    # Tokenize the entire description
    tokens = tokenizer.tokenize(description)
    # print(len(tokens), torch.sum(tokenizer(raw_descriptions, padding=True, truncation=True, return_tensors="pt")['input_ids'][i]!=0))
    words = description.split()

    current_char_index = 0
    word_index = 0
    word_start = 0
    word_end = len(words[word_index])

    word_token_map = []

    for token in tokens:
        while current_char_index < len(description) and description[current_char_index].isspace():
            current_char_index += 1  # Skip any whitespace characters

        token_start = current_char_index
        token_end = token_start + len(token.replace("##", ""))

        if token_end > word_end:
            word_index += 1
            word_start = word_end + 1
            word_end = word_start + len(words[word_index])

        word_token_map.append((token, word_index))
        current_char_index = token_end

    token_to_word_index_mappings.append(word_token_map)
tokens_labels = []
tokens_logits = []
# Print the token to word index mapping for each description
for i, description in enumerate(raw_descriptions):
  # 0 at the begining for bert [cls]
    tokens_labels_i = [0.0]
    tokens_logits_i = [0.0]
    print(f"Description: {description}")
    print("Token to Word Index Mapping:")
    for token, word_index in token_to_word_index_mappings[i]:
        print(f"Token: {token} - Word Index: {word_index} - Word {description.split()[word_index]}")
        print(f"Token: {token} - label: {words_labels[i][word_index]}, prob: {words_probs[i][word_index]}")
        tokens_labels_i.append(words_labels[i][word_index])
        tokens_logits_i.append(words_probs[i][word_index])
  # 0 at the end for bert [eos]
    tokens_labels_i.append(0.0)
    tokens_logits_i.append(0.0)

    tokens_labels.append(tokens_labels_i)
    tokens_logits.append(tokens_logits_i)
```

#Train

```python
import os
import random
import torch
from torch.utils.data import Dataset, DataLoader, random_split
from transformers import AutoTokenizer, BertForTokenClassification, BertConfig
from tqdm import tqdm
import numpy as np

# Set seeds
def set_seed(seed):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    # For some operations to ensure deterministic behavior
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

# Set a specific seed
seed = 42
set_seed(seed)

# Tokenizer
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

def encode_data(raw_texts, labels, features):
    tokenized_texts = tokenizer(raw_texts, padding=True, truncation=True, return_tensors="pt")
    input_ids = tokenized_texts['input_ids']
    attention_mask = tokenized_texts['attention_mask']

    max_len = input_ids.shape[1]
    aligned_labels = []
    aligned_features = []

    for i, label_list in enumerate(labels):
        aligned_label = [-100] * max_len
        aligned_feature = [0.0] * max_len

        for j in range(len(label_list)):
            aligned_label[j] = label_list[j]
            aligned_feature[j] = features[i][j]
        # print(len(label_list), len(raw_texts[i].split()))

        aligned_labels.append(aligned_label)
        aligned_features.append(aligned_feature)


    clipped_aligned_features = torch.clamp(torch.tensor(aligned_features), min=0, max=1)
    return input_ids, attention_mask, torch.tensor(aligned_labels, dtype=torch.float32), clipped_aligned_features

input_ids, attention_mask, labels, features = encode_data(raw_descriptions, tokens_labels, tokens_logits)

# Check the shapes and types of tensors
print(f"Input IDs shape: {input_ids.shape}, dtype: {input_ids.dtype}")
print(f"Attention Mask shape: {attention_mask.shape}, dtype: {attention_mask.dtype}")
print(f"Labels shape: {labels.shape}, dtype: {labels.dtype}")
print(f"Features shape: {features.shape}, dtype: {features.dtype}")

class TokenClassificationDataset(Dataset):
    def __init__(self, input_ids, attention_mask, labels, features):
        self.input_ids = input_ids
        self.attention_mask = attention_mask
        self.labels = labels
        self.features = features

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        item = {
            'input_ids': self.input_ids[idx],
            'attention_mask': self.attention_mask[idx],
            'labels': self.labels[idx],
            'features': self.features[idx]
        }
        return item

dataset = TokenClassificationDataset(input_ids, attention_mask, labels, features)

# Split the dataset into training and testing sets (80-20 split)
train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size
train_dataset, test_dataset = random_split(dataset, [train_size, test_size])

# Create dataloaders
train_dataloader = DataLoader(train_dataset, batch_size=1, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=1, shuffle=False)

class CustomBertForTokenClassification(BertForTokenClassification):
    def __init__(self, config):
        super().__init__(config)
        self.classifier = torch.nn.Sequential(
            torch.nn.Linear(config.hidden_size + 1, (config.hidden_size + 1)),
            torch.nn.ReLU(),
            torch.nn.Linear((config.hidden_size + 1), 1)
        )

    def forward(self, input_ids=None, attention_mask=None, token_type_ids=None, labels=None, features=None):
        device = self.bert.embeddings.word_embeddings.weight.device
        input_ids = input_ids.to(device)
        attention_mask = attention_mask.to(device)
        if token_type_ids is not None:
            token_type_ids = token_type_ids.to(device)
        if labels is not None:
            labels = labels.to(device)
        if features is not None:
            features = features.to(device)

        outputs = self.bert(input_ids, attention_mask=attention_mask, token_type_ids=token_type_ids)
        sequence_output = outputs[0]
        extended_output = torch.cat((sequence_output, features.unsqueeze(-1)), dim=-1)
        logits = self.classifier(extended_output)
        logits = torch.sigmoid(logits)

        loss = None
        if labels is not None:
            loss_fct = torch.nn.BCEWithLogitsLoss()
            loss_fct = torch.nn.CrossEntropyLoss(reduction='none')  # Use reduction='none' to get per-token loss
            # Calculate loss, masking out tokens with attention_mask=0
            active_loss = labels.view(-1) != -100  # Ignore padding tokens
            active_logits = logits.view(-1)[active_loss]
            active_labels = labels.view(-1)[active_loss]
            # print(active_logits)
            # print(active_labels)
            loss = loss_fct(active_logits, active_labels)
            loss = loss.mean()
        output = (logits,) + outputs[2:]
        return ((loss,) + output) if loss is not None else output

# Define model configuration
config = BertConfig.from_pretrained('bert-base-uncased', num_labels=2)
model = CustomBertForTokenClassification.from_pretrained('bert-base-uncased', config=config)

# Move model to GPU
device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
model.to(device)

# Optimizer
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)

# Training parameters
epochs = 3

# Training loop
model.train()
for epoch in range(epochs):
    print(f'Epoch {epoch + 1}/{epochs}')
    epoch_loss = 0
    for batch in tqdm(train_dataloader):
        optimizer.zero_grad()

        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)
        features = batch['features'].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels, features=features)
        loss = outputs[0]
        print(f"loss is {loss.item()}")

        loss.backward()
        optimizer.step()

        epoch_loss += loss.item()

    print(f'Loss: {epoch_loss / len(train_dataloader)}')

print("Training complete")
```

```python
from functools import reduce

def flatten_list(lst):
    return reduce(lambda x, y: x + (flatten_list(y) if isinstance(y, list) else [y]), lst, [])
```

```python
from sklearn.metrics import f1_score
from tqdm import tqdm

# Move model to CPU
model.to(torch.device('cpu'))

# Predict on the entire data
predictions = []
true_labels = []

for batch in tqdm(test_dataloader):
    input_ids = batch['input_ids'].to(torch.device('cpu'))
    attention_mask = batch['attention_mask'].to(torch.device('cpu'))
    labels = batch['labels'].to(torch.device('cpu'))
    features = batch['features'].to(torch.device('cpu'))

    outputs = model(input_ids=input_ids, attention_mask=attention_mask, features=features)
    logits = outputs[0]
    threshold = 0.5
    active_logits = logits[labels != -100]
    active_labels = labels[labels != -100]
    curr_pred = torch.where(active_logits > threshold, torch.tensor(1), torch.tensor(0)).tolist()
    # print(active_logits, active_labels)
    predictions.append(curr_pred)
    true_labels.append(active_labels.tolist())

    input_ids_list = input_ids.tolist()
    # print(input_ids_list)
    # Decode each list of input_ids to a string
    pred_flattened = (flatten_list(curr_pred))
    labels_flattened = flatten_list((active_labels.tolist()))
    tokens_list = flatten_list([tokenizer.convert_ids_to_tokens(input_ids) for input_ids in input_ids_list])
    tokens_pred_tuples = []
    for i in range(len(pred_flattened)):
      tokens_pred_tuples.append((tokens_list[i], pred_flattened[i], labels_flattened[i]))
    # print(tokens_pred_tuples)
    html_text = []
    for word, pred, gt in tokens_pred_tuples:
      if pred == 1 and gt == 0.0:
          html_text.append(get_colored_text(word, 'red'))
      elif pred == 1 and gt == 1.0:
          html_text.append(get_colored_text(word, 'green'))
      elif gt == 1.0 and pred == 0:
          html_text.append(get_colored_text(word, 'yellow'))
      else:
          html_text.append(word)
    # Join the HTML formatted text
    html_output = ' '.join(html_text)

    # Display the HTML formatted text
    display(HTML(html_output))


# Flatten the lists
predictions = [item for sublist in predictions for item in sublist]
true_labels = [item for sublist in true_labels for item in sublist]

# Compute F1 score
f1 = f1_score(true_labels, predictions, average='macro')

print(f"F1 score: {f1}")
# prompt: add confusion matrix and print it pretty

from sklearn.metrics import confusion_matrix
import pandas as pd
import numpy as np

# Find unique labels
unique_labels = [ 0, 1]

# Compute confusion matrix
cm = confusion_matrix(true_labels, predictions, labels=unique_labels)

# Create pandas dataframe for pretty printing
df_cm = pd.DataFrame(cm,
                     columns=['Predicted ' + str(label) for label in unique_labels],
                     index=['Actual ' + str(label) for label in unique_labels])

# Print confusion matrix
print(df_cm.to_string())
```
